```python
import csv 
import json
import pandas as pd
```

```python
with open('data_raw.json') as f:
    data = json.load(f)
```

```python
# Converting the data into datafrome
df = pd.DataFrame(data)
```

```python
# Transformation 
# adding inr price also
usd_to_inr = 88.04
df['price_inr'] = (df['price_usd'] * usd_to_inr).round(2)

# Adding tax_inr column also for tax
df['tax_inr'] = (df['price_inr'] * 0.18).round(2) # 18% GST
```

```python
# Renaming column "Name" to "Item"
df = df.rename(columns={
    'name' : 'item'
})
```

```python
df_final = df[['id', 'item', 'category', 'price_usd', 'price_inr', 'tax_inr', 'rating']]
df_final
```

```python
# Loading json transformend data into csv file 
df_final.to_csv('products.csv', index=False)
print("✅ Transformation complete → products.csv generated successfully!")
df_final.head()
```
